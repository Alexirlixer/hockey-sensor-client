```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
cols = ['ts', 'gx', 'gy', 'gz', 'ax', 'ay', 'az']
df = pd.read_csv('./data-radar-snap.csv', names = cols)
```

```python
df['acc'] = np.sqrt(df['ax']**2 + df['ay']**2 + df['az']**2)
```

```python
fig = plt.figure(figsize = (12, 4))
sns.lineplot(df['acc'])
fig.show()
```
