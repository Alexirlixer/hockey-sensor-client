```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
cols = ['ts', 'gx', 'gy', 'gz', 'ax', 'ay', 'az']
dfs = pd.read_csv('data-static.csv', names=cols)
dfm = pd.read_csv('data-move.csv', names=cols)
```

```python
fig, axes = plt.subplots(nrows=6, ncols=1, figsize=(30,40))

i = 0
for name in ['ax', 'ay', 'az']:
    sns.histplot(data=dfs[name],ax=axes[i])
    axes[i].set_title(f"static {name} plot")
    sns.histplot(data=dfm[name],ax=axes[i+1])
    axes[i+1].set_title(f"move {name} plot")
    i += 2
plt.show()
```

```python
fig, axes = plt.subplots(nrows=6, ncols=1, figsize=(30,40))

i = 0
for name in ['gx', 'gy', 'gz']:
    sns.histplot(data=dfs[name],ax=axes[i])
    axes[i].set_title(f"static {name} plot")
    sns.histplot(data=dfm[name],ax=axes[i+1])
    axes[i+1].set_title(f"move {name} plot")
    i += 2
plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(30, 20))

sacc = np.sqrt(dfs['ax']**2 + dfs['ay']**2 + dfs['az'] ** 2)
macc = np.sqrt(dfm['ax']**2 + dfm['ay']**2 + dfm['az'] ** 2)

sns.lineplot(data=sacc, ax=axes[0])
axes[0].set_title('static acceleration')

sns.lineplot(data=macc, ax=axes[1])
axes[1].set_title('move acceleration')

plt.show()
```
