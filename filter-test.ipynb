```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
cols = ['ts', 'gx', 'gy', 'gz', 'ax', 'ay', 'az']
dft = pd.read_csv('data-test.csv', names=cols)
```

```python
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(30,30))

i = 0
for name in ['ax', 'ay', 'az']:
    sns.histplot(data=dft[name],ax=axes[i])
    axes[i].set_title(f"test {name} plot")
    i += 1
plt.show()
```

```python
acc = np.sqrt(dft['ax']**2 + dft['ay']**2 + dft['az'] ** 2)

plt.figure(figsize=(30,10))
sns.lineplot(data=acc)
plt.show()
```
